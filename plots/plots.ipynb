```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('CTD2_raw_output.xlsx')
print(df.head())
print(df.shape)
```

```python
alpha = df['ALPHA']
display(alpha.head())

beta = df['BETA']
display(beta.head())
```

```python
new_df = pd.concat([alpha, beta], axis=1)
display(new_df.head())
```

```python
df_with_labels = pd.read_csv('new_df.csv')
display(df_with_labels.head())
```

```python
alpha = df_with_labels['ALPHA']
beta = df_with_labels['BETA']
labels = df_with_labels['LABELS']
```

```python
type_label = []
for label in labels:
    label_type = label.split('_')[1]
    type_label.append(label_type)
```

```python
df_with_labels['LABELS'] = type_label
display(df_with_labels.head())
df_with_labels.to_csv('new_df.csv', index=False)
```
