```python
import pandas as pd
from pycaret.regression import predict_model
from pycaret.regression import load_model
import os
```

```python
df_NCI = pd.read_csv('data/NCI_60ab.csv', index_col=0)
df_NCI = df_NCI[['Alpha', 'Beta']]
df_NCI.head()
```

```python
df_entrez = pd.read_csv('data/entrez_og.csv')
df_entrez.head()
```

```python
df_entrez['cell_line'] = df_entrez['cell_line'].str.replace('-2', '')
df_entrez['cell_line'] = df_entrez['cell_line'].str.replace('-5', '')
df_entrez['cell_line'] = df_entrez['cell_line'].str.replace('-8', '')
```

```python
df_entrez.head()
```

```python
df_entrez.drop_duplicates(subset=['cell_line'], inplace=True)
df_entrez.reset_index(drop=True, inplace=True)
df_entrez.set_index('cell_line', inplace=True)
df_entrez.head()
```

```python
alpha_model = load_model('models/0.86MLKing6302022')
beta_model = load_model('models/finalBetaModel1200features')
```

```python
alpha_predicted = predict_model(alpha_model, data=df_entrez, round=0)
alpha_predicted = alpha_predicted['Label']
renamed_alpha = alpha_predicted.rename("alpha_predicted")
renamed_alpha.head()
```
