```python
from pycaret.regression import *
import pandas as pd
# check version
from pycaret.utils import version
version()
```

```python
dataset = pd.read_csv('data/only2_big_with_sf.csv')
dataset.set_index('cell_line', inplace=True)
dataset.head()
```

```python
dataset.shape
```

```python
data = dataset.sample(frac=0.9, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions ' + str(data_unseen.shape))
```

```python
from pycaret.regression import *
```

```python
exp_reg102 = setup(data = data, target = 'SF', session_id=123,
                  normalize = True, transformation = True, transform_target = True,
                  combine_rare_levels = True, rare_level_threshold = 0.05,
                  remove_multicollinearity = True, multicollinearity_threshold = 0.95,
                  bin_numeric_features = ['1462', '200916','254158'],
                  log_experiment = True, experiment_name = 'sf1')
```
