```python
import pandas as pd
from sklearn.model_selection import train_test_split
from PyImpetus import PPIMBR
```

```python
ccleRNAEnsemble = pd.read_csv('ccle_expressionensemble_Entrez_mergedwAB.csv', index_col=0)
ccleRNAEnsemble.head()
```

```python
X_train, X_test = train_test_split(ccleRNAEnsemble, test_size = 0.10, random_state=7)
X_train = X_train.drop('X', axis=1)
X_train.head()
```

```python
X_trainAlpha = X_train.drop(['Alpha', 'Beta'], axis=1)
X_trainAlpha.head()
```

```python
alpha_vals = X_train['Alpha'].values
alpha_vals
```

```python
from sklearn.linear_model import BayesianRidge

model = PPIMBR(model=BayesianRidge(), p_val_thresh=0.05, num_simul=50, simul_size=0.2,sig_test_type="non-parametric", random_state=27, n_jobs=-1, verbose=0)
# The fit_transform function is a wrapper for the fit and transform functions, individually.
# The fit function finds the MB for given data while transform function provides the pruned form of the dataset
df_train = model.fit_transform(X_trainAlpha, alpha_vals)
# Check out the MB
print(model.MB)
```
