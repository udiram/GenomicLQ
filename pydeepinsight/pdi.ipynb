```python
X_trainCCLER = pd.read_csv('ccle_rnaseq_deseq_23456.csv', index_col=0)
X_trainCCLES = pd.read_csv('ccle_rnaseq_deseq_23456_structured.csv', index_col=0)
X_trainCCLEY = pd.read_csv('ccle_rnaseq_deseq_23456_SForY.csv', index_col=0)
```

```python
scaler = MinMaxScaler()
X_train_normR = scaler.fit_transform(X_trainCCLER)
```

```python
reducer = umap.UMAP(n_neighbors=200)
```

```python
it = ImageTransformer(pixels=40, feature_extractor=reducer)
X_train_imgR = it.fit_transform(X_train_normR)
```

```python
from matplotlib import pyplot as plt
for enum, image in enumerate(X_train_imgR):
     try:
        # save image
        plt.imsave('images_og/image_{}.png'.format(name_list[enum]), image)
     except:
        # skip the current image if there's an error
        print('Skipping image_{}.png due to an error.'.format(name_list[enum]))
        continue
```

```python
import os
import cv2
print(len(os.listdir('images_og')))
for image_name in os.listdir('images_og'):
    print(image_name)
    ending = image_name.split('-')[-1].split('.')[0]
    print(ending)
    image = cv2.imread('images_og/' + image_name)
    # print(image)
    if ending == '2':
        dir = '2/'
        r_rect = np.zeros_like(image)
        mask = r_rect[:, :, 2] = 255
        r_rect_masked = cv2.bitwise_and(image, r_rect)
        cv2.imwrite(dir + image_name, r_rect_masked)
    elif ending == '3':
        dir = '3/'
        r_rect = np.zeros_like(image)
        mask = r_rect[:, :, 1] = 255
        r_rect_masked = cv2.bitwise_and(image, r_rect)
        cv2.imwrite(dir + image_name, r_rect_masked)
    elif ending == '4':
        dir = '4/'
        r_rect = np.zeros_like(image)
        mask = r_rect[:, :, 0] = 255
        r_rect_masked = cv2.bitwise_and(image, r_rect)
        cv2.imwrite(dir + image_name, r_rect_masked)
    elif ending == '5':
        dir = '5/'
        r_rect = np.zeros_like(image)
        mask = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
        mask[:, :, 0] = 255  # Set blue channel to 255
        mask[:, :, 1] = 255  # Set green channel to 255
        mask[:, :, 2] = 0    # Set red channel to 0
        r_rect_masked = cv2.bitwise_and(image, mask)
        cv2.imwrite(dir + image_name, r_rect_masked)
    else:
        dir = '6/'
        r_rect = np.zeros_like(image)
        mask = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
        mask[:, :, 0] = 0    # Set blue channel to 0
        mask[:, :, 1] = 255  # Set green channel to 128
        mask[:, :, 2] = 255  # Set red channel to 255
        r_rect_masked = cv2.bitwise_and(image, mask)
        cv2.imwrite(dir + image_name, r_rect_masked)
```
