```python
import pandas as pd
```

```python
df_merged = pd.read_csv('data/ccleRSEMTPMmergedwAB.csv')
df_merged.rename(columns={'Row.names':'CCLE_Name'}, inplace=True)
df_merged.head(50)
```

```python
df_og = pd.read_csv('data/CCLE_expressionEnsemble_DepMap.csv', index_col=0)
df_og.head(50)
renamed_list = []
for i in df_og['CCLE_Name']:
    print(i.split('_')[0])
    renamed_list.append(i.split('_')[0])
df_og['CCLE_Name'] = renamed_list
df_og.head(50)
```

```python
df_merged_2 = pd.merge(df_og, df_merged, on='CCLE_Name')
df_merged_2.head(50)
```

```python
df_merged_2.to_csv('data/ccleRSEMTPMmergedwAB_DepMap.csv')
```

```python
df_dep_map = pd.read_csv('data/CCLE_expressionEnsemble_DepMap.csv', index_col=0)
df_dep_map.head(50)
```

```python
# take columns alpha and beta from merged_2 and merge with dep_map
alpha_beta = df_merged_2[['Alpha', 'Beta']]
df_dep_map['alpha'] = alpha_beta['Alpha']
df_dep_map['beta'] = alpha_beta['Beta']
df_dep_map.head(50)
```

```python
df_dep_map.to_csv('data/ccleRSEMTPMmergedwAB_DepMap_alpha_beta.csv')
```
